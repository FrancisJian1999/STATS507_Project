```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
from pathlib import Path
```

```python
data = pd.read_csv(Path('yjy_data.csv'), index_col=0)
```

```python
# Replace empty data and 8/9 data
data = data.replace(r'^\s*$', pd.np.nan, regex=True)
data.dropna(how="any", inplace=True)

# Replace Survey Values with Dictionary Values
# data.replace({"RATEX" : ratex_dict}, inplace=True)

data = data.drop_duplicates()

data = data.reset_index(drop=True)
data = data.astype(float)

data.replace({'PJOB': 998}, pd.np.nan, inplace=True)
data.replace({'PJOB': 999}, pd.np.nan, inplace=True)
data.dropna(how="any", inplace=True)

data.dtypes
```

```python
# fig = px.scatter_matrix(data)
# fig.show()

# plot a scatterplot of column1 vs column2
plt.scatter(data['INCOME'], data['PJOB'])

# add labels and title
plt.xlabel('Income Expectation')
plt.ylabel('Unemployment Expectation')
plt.title('"Income Expectation vs Unemployment Expectation"')
```
